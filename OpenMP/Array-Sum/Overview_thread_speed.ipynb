# Overview thread speed

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df_data = pd.read_csv('overview-speedtest.csv')
df_data.set_index('Unnamed: 0')
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(15, 7))

# plot for every thread the speed
x = df_data.columns[1:]

for hue in df_data.values:
    plt.plot(x, hue[1:], label=hue[0])

# add legend
plt.legend(loc="upper left")

# add title and x, y labels
plt.title("Snelheid array sum functie per hoeveelheid beschikbare threads")
plt.xlabel("Lengte lijst met getallen")
plt.ylabel("Snelheid (sec)")

plt.show()
```
