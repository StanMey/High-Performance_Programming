```python
# Computes an approximation of Pi by summing up
# a series of terms.  The more terms, the closer
# the approximation.
# Based on a variation of the Gregory-Leibniz series approximation
#
import threading
import logging
import math
import numpy as np


def f( x ):
    """
    The function being integrated, and which returns
    an approximation of Pi when summed up over an interval
    of integers.
    """
    return 4.0 / ( 1 + x*x )

class WorkerThread(threading.Thread):
    
    def __init__(self, args):
        """
        args must contain n1, n2, and deltaX, in that order.
        """
        threading.Thread.__init__(self, args=args)
        self.n1 = args[0]
        self.n2 = args[1]
        self.deltaX = args[2]
        self.calc_sum = 0.0
    
    def run(self):
        """
        This will be called automatically by start() in
        main.  It's the method that does all the work.
        """
        for x in range(self.n1, self.n2):
            self.calc_sum += f(x * self.deltaX)
    
    def get(self):
        """
        gets the computed sum.
        """
        return self.calc_sum

if __name__ == "__main__":
    format = "%(asctime)s: %(message)s"
    logging.basicConfig(format=format, level=logging.INFO,
                        datefmt="%H:%M:%S")
    
    # Gets a number of terms from the user, then sums
    # up the series and prints the resulting sum, which
    # should approximate Pi.
    
    
    thread_amount = 3
    
    # get the number of terms
    N = int( input( "> " ) )

    sum = 0.0          # where we sum up the individual
                       # intervals
    deltaX = 1.0 / N   # the interval
    
    # get the ranges of the array to select
    ranges = list()
    n1 = 0
    delta = math.ceil(N/thread_amount)
    for i in range(thread_amount):
        n2 = min(n1 + delta, N)
        ranges.append((n1, n2))
        n1 = n2
    
    # create the array with numbers
    num_array = [x for x in range(N)]
    
    # 
    threads = list()
    for index in range(thread_amount):
        logging.info("Main    : create and start thread %d.", index)
        n_range = ranges[index]
        x = WorkerThread([n_range[0], n_range[1], deltaX])
        threads.append(x)
        x.start()
    
    for index, thread in enumerate(threads):
        logging.info("Main    : before joining thread %d.", index)
        thread.join()
        sum += thread.calc_sum
        logging.info("Main    : thread %d done", index)
    
    # display results
    print( "sum = %1.9f" % ( sum * deltaX ) )
```
